```
a = [111,222,333]
for i, idx in enumerate(a):
    print (i)
    print (idx)
```

```
def count_ups(data):
    count = 0
    for idx,d in enumerate(data):
        if idx == 0: # if first loop, skip.
            continue
        if d > data[idx-1]: # If current is greater than last
            count +=1
    return count
```
