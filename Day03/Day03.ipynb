```
from scipy.stats import mode
#with open('Day03.example','r') as f:
with open('Day03.input','r') as f:
    rawdata = f.read().splitlines()
    data = []
    for r in rawdata:
        data.append([int(char) for char in r])
print(data)

```

# Part 1
## Return column as list

```
def column(col, data):
    ans = []
    for l in data:
        for idx, value in enumerate(l):
            if idx == col:
                ans.append(value)
    return ans
```

## Return Most Frequent

```
def most_freq(line):
    return mode(line)[0][0]
```

## Loop through input, return gamma (also epsilon)

```
def gamma(data):
    gamma = []
    for idx in range(len(data[0])):  # for number of columns
        gamma.append(most_freq(column(idx,data)))
    gamma_str = [str(q) for q in gamma]
    gamma = ''.join(gamma_str)
    epsilon = ''.join('1' if x == '0' else '0' for x in gamma_str)
    return int(gamma,2), int(epsilon,2)
```

## Run it

```
g, e = gamma(data)
print('Answer part 1: {}'.format(g*e))
```
