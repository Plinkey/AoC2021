```
import numpy as np
#with open('Day05.example','r') as f:
with open('Day05.input','r') as f:
    rawdata = f.read().splitlines()
    #rawdata = rawdata[0].split(' -> ')
    data = []
    for l in rawdata:
        t = l.split(' -> ')
        a = [int(i) for i in t[0].split(',')]
        b = [int(i) for i in t[1].split(',')]
        data.append([a,b])

field = np.zeros([1000,1000])
```

```
def draw(start, stop):
    for x in range(start[0], stop[0]+1) or range(start[0], stop[0]-1, -1):
        for y in range(start[1], stop[1]+1) or range(start[1], stop[1]-1, -1):
            print('X: {}, Y: {}'.format(x,y))
            field[y,x] += 1
```

```
for line in data:
    start = line[0]
    stop = line[1]
    if start[0] == stop[0]:
        print('Running line Start: {}, Stop: {}'.format(start,stop))
        draw(start, stop)
    elif start[1] == stop[1]:
        print('Running line Start: {}, Stop: {}'.format(start,stop))
        draw(start, stop)
```

```
count = np.count_nonzero(field >= 2)
print(count)
```
