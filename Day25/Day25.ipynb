```python
with open('Day25.input','r') as f:
#with open('Day25.example','r') as f:
    rawdata = f.read().splitlines()

field = []
for line in rawdata:
    tmp = [char for char in line]
    field.append(tmp)

fieldH = len(field[0])
fieldV = len(field)
```

# HELPER FUNCTIONS

```python
def reset():
    field = []
    for line in rawdata:
        tmp = [char for char in line]
        field.append(tmp)
    return field
```

```python
def move_east(v,h,f):
    dest = h + 1
    if h == fieldH-1: #Last column, moves to first
        dest = 0
    f[v][h] = '.'
    f[v][dest] = '>'
    return f
```

```python
def move_south(v,h,f):
    dest = v + 1
    if v == fieldV-1: #Last row, moves to first
        dest = 0
    f[v][h] = '.'
    f[dest][h] = 'v'
    return f
```

```python
def check_no_neighbor(v,h):
    if field[v][h] == '>':
        dest = h+1
        if h == fieldH-1: #Last column moves to first
            dest = 0
        if field[v][dest] == '.':
            return True
        else:
            return False
    elif field[v][h] == 'v':
        dest = v+1
        if v == fieldV - 1: # Last row, moves to first
            dest = 0
        if field[dest][h] == '.':
            return True
        else:
            return False
```

```python
def print_field():
    for line in field:
        print(''.join(line))
```

```python
field = reset()
print_field()
```

```python
def next_step_easts():
    no_move_flag = True
    new = copy.deepcopy(field)
    for vdx in range(fieldV):
        for hdx in range(fieldH):
            if check_no_neighbor(vdx, hdx):
                if field[vdx][hdx] == '>':
                    move_east(vdx, hdx, new)
                    no_move_flag = False
    return new, no_move_flag
```

```python
def next_step_souths():
    no_move_flag = True
    new = copy.deepcopy(field)
    for vdx in range(fieldV):
        for hdx in range(fieldH):
            if check_no_neighbor(vdx, hdx):
                if field[vdx][hdx] == 'v':
                    move_south(vdx, hdx, new)
                    no_move_flag = False
    return new, no_move_flag
```

# Testing, run only a certain number of steps

```python
nSteps = 50
field = reset()
for _ in range(nSteps):
    field, flag = next_step_easts()
    field, flag = next_step_souths()

#print_field()
```

# Run Part One

```python
field = reset()
nRounds = 0
while True:
    nRounds += 1
    field, east_flag = next_step_easts()
    field, south_flag = next_step_souths()
    if east_flag and south_flag:
        break

print('Answer to part one is: {}'.format(nRounds))
```
